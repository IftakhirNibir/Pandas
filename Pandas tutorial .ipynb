## Loading data into Pandas 

```python
import pandas as pd 
df = pd.read_csv("pokemon_data.csv")
```

## Reading data in Pandas 

```python
#Read Header
df.columns 
```

```python
#Read each column
df.Name  #All values in 'Name' column
df.Name[:4] #First five values in 'Name' column
df['HP'] #All values in 'HP' column
df[['Name','Type 1','HP']] #Showing these three columns with values 
```

```python
#Read each row
df.head(3) #Showing first 3 rows 
#iloc standers for Integer location
df.iloc[1] #It show the first rows [header row]
#Read a specific location using iloc[row,column]
#Q:Find this value 'Bulbasaur' using iloc concept 
df.iloc[0,1]
```

```python
#loc is use for finding specific data in our data set
df.loc[df['Type 1']=='Fire'][:3]
```

## Sorting / Describing data

```python
#Describing data 
df.describe()
```

## Making changes to the data

```python
#if we want to make changes in horizontally then we need to use 'axis=1'
#if vertically we need to use 'axis=0'
#Making a new column which returns total 
df['Total'] = df['HP']+df['Attack']+df['Defense']+df['Sp. Atk']+df['Sp. Def']+df['Speed']+df['Generation']
#Another way
df['Total'] = df.iloc[:,4:10].sum(axis=1)
```

```python
df.head()
```

```python
# dropping a column
df = df.drop(columns=['Total'])
```

```python
df.head()
```

## Saving our Data(Exporting into Desired Format) 

```python
#index=False meaning not showing default indexing value
#Saving file in the csv formate
df.to_csv('CSV.csv',index=False)
#excel formate
df.to_excel('Excel.xlsx',index=False)
#text formate
df.to_csv('Text.txt',index=False,sep='\t')
#sep='\t' means separate by tab
```

## Filtering Data

```python
df.loc[(df['Type 1']=='Grass')&(df['Type 2']=='Poison')&(df['HP']>75)]
```

```python
#if we want to remove unordered default indexing and save
new_df = df.loc[(df['Type 1']=='Grass')&(df['Type 2']=='Poison')&(df['HP']>75)]
#new_df = new_df.reset_index() #it makes an ordered indexing but also showing
# the default ordered column.To remove this we need to use 'drop=True'
#new_df = new_df.reset_index(drop=True)
#if we don't want to have to reset it to a new dataframe we need to use
#'inplace=True' So
new_df.reset_index(drop=True,inplace=True)
new_df
```

```python
#Find a value which contain 'Mega' inside 'Name' column
df.loc[df['Name'].str.contains('Mega')]
```

```python
#if we want to do the opposite operation 
df.loc[~df['Name'].str.contains('Mega')]
```

```python
#Q:Find all values in 'Type 1' column where it contains either 'Fire' or 'Grass'
df.loc[df['Type 1'].str.contains('Fire|Grass')]
#If we want to ignore the case sensitivity 
import re
df.loc[df['Type 1'].str.contains('fire|grass',flags=re.I)]
```

```python
#Q:Find those names which is start at 'Pi' and contains any letter and then 
#contains zero or more letter
kdf = df.loc[df['Name'].str.contains('^Pi[a-z]*',flags=re.I)]
kdf.reset_index(drop=True,inplace=True)
kdf
```

## Conditional Changes

```python
#replace column values
df.head()
#Q:Replace 'Type 1' column 'Grass' to 'Something'
df.loc[df['Type 1']=='Grass','Type 1'] = 'Something'
df.head()
```

```python
#Q:If 'Type 2' column is 'Poison' then change 'Legendary' column = True
df.loc[df['Type 2']=='Poison','Legendary'] = True
df.head()
```

```python
#We can also change multiple column values 
df.loc[df['HP']>=45,['Generation','Legendary']] = 'Test Value'
df.head()
```

```python
#we can also changes multiple column value differently 
df.loc[df['HP']>=45,['Generation','Legendary']] = ['Test 1','Test 2']
df.head()
```

## Aggregate Statistics (Groupby)

```python
df = pd.read_csv('CSV.csv')
df.head()
df.groupby('Type 1').sum().sort_values('Defense',ascending=False)
#like sum() we can also find mean(),count()
```

```python
#We can use multiple columns in groupby clause
df['count'] = 1
df.groupby(['Type 1','Type 2']).count()['count']
```
